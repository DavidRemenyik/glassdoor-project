# Calculating Category Weights

With the cleaned datasets, we can now calculate the "weights" for each category we will use to determine our own overall rating. We calculate this via the following simple algorithm:

```python
for category, pattern in categories.items():
  pros = chunk["pros"].str.contains(pattern, regex=True).sum()
  cons = chunk["cons"].str.contains(pattern, regex=True).sum()
  
  weights[category] += pros - cons
```

Essentially, whenever we come across a word related to a given category in the "pros" column, we add +1 to that category's score. Similarly, if we find a word associated with a given category in the "cons" category, we subtract -1 from that category's score.

````python
!time python3 python/calculate-weights.py dataset1.csv
````

````python
!time python3 python/calculate-weights.py dataset2.csv
````

From running the script on both datasets, we get the following total weights which we can use to weight our own rating system when we execute our SQL queries:


| Category           | dataset1 | dataset2 | Total |
| --------           | --- | --- | --- |

To convert these values to more appropriate floating point values:

$$
\text{weight_category} = total / \sum |total| \\ \\
|total| = 315929 + 1121671 + 13820 + 34236 + 682293 + 883812 = 2,639,761 \\
$$

````python
# The firms with the highest rating (with at least 1000 reviews)
!time mysql -e " \
USE glassdoorReviews; SELECT \
    firm, \
    AVG(overall_rating) AS avg_rating, \
    COUNT(*) AS review_count \
FROM employee \
WHERE overall_rating IS NOT NULL \
GROUP BY firm \
HAVING COUNT(*) >= 1000 \
ORDER BY avg_rating DESC \
LIMIT 10; \
"
````

````python
# The firms with the most reviews and their average ratings
!time mysql -e " \
USE glassdoorReviews; SELECT \
    firm, \
    COUNT(*) AS review_count, \
    AVG(overall_rating) AS avg_rating \
FROM employee \
WHERE overall_rating IS NOT NULL \
GROUP BY firm \
ORDER BY review_count DESC \
LIMIT 10; \
"
````

````python
# The firms with the most reviews and their average ratings (with our weights)
!time mysql -e " \
USE glassdoorReviews; WITH top10 AS ( \
    SELECT firm \
    FROM employee \
    GROUP BY firm \
    ORDER BY COUNT(*) DESC \
    LIMIT 10 \
) \
SELECT \
    e.firm, \
    COUNT(*) AS review_count, \
    (AVG(\
          e.work_life_balance   *  1.1197 \
        + e.culture_values      *  1.4247 \
        + (e.career_opp          + (e.career_opp * -0.0129)) \
        + e.comp_benefits       *  1.2584 \
        + (e.senior_mgmt         + (e.senior_mgmt * -0.3348)) \
    ) / 5) AS weighted_overall_rating \
FROM employee e \
JOIN top10 t ON e.firm = t.firm \
GROUP BY e.firm \
ORDER BY review_count DESC; \
"
````

````python
# The firms with the highest rating (with at least 1000 reviews, with our weights)
!time mysql -e " \
USE glassdoorReviews; \
SELECT \
    firm, \
    COUNT(*) AS review_count, \
    (AVG(\
          work_life_balance   *  1.1197 \
        + culture_values      *  1.4247 \
        + (career_opp          + (career_opp * -0.0129)) \
        + comp_benefits       *  1.2584 \
        + (senior_mgmt         + (senior_mgmt * -0.3348)) \
    ) / 5) AS weighted_rating \
FROM employee \
WHERE work_life_balance IS NOT NULL \
  AND culture_values IS NOT NULL \
  AND career_opp IS NOT NULL \
  AND comp_benefits IS NOT NULL \
  AND senior_mgmt IS NOT NULL \
GROUP BY firm \
HAVING COUNT(*) >= 1000 \
ORDER BY weighted_rating DESC \
LIMIT 10; \
"
````
